### Task 3.2

```python
wineData = np.load('Datas/wineData.npy')
wineLabels = np.load('Datas/wineLabels.npy')

print("The shape of wineData is ",wineData.shape)
print("The shape of wineLabels is ",wineLabels.shape)

plt.figure()

plt.scatter(wineData[:,0],
            wineData[:,1],
            c = wineLabels)
plt.show()
```

The data that we're using consists of 178 elements that have 13 features.

```python
# Splicing the data into 145 and 33 elements (80% training 20% testing)
train_data = wineData[:145]
test_data = wineData[145:]
# Splicing the labels into 145 and 33 elements (80% training 20% testing)
train_labels = wineLabels[:145]
test_labels = wineLabels[145:]
```

```python
# Defining the function to scatter training and testing datas
def scatter(trainData, testData, trainLabels, testLabels):
    # scatter the Dimensionally reduced datas
    plt.figure()
    # first the training data
    plt.scatter(trainData[:,0],
                trainData[:,1],
                c = trainLabels)
    # then the test data
    plt.scatter(testData[:,0],
                testData[:,1],
                c = testLabels,
                marker = "x")



scatter(train_data, test_data, train_labels, test_labels)
plt.title('Wine dataset in original feature space')
plt.xlabel('Original feature 0')
plt.ylabel('Original feature 1')
plt.show()
```

Scatter plot of the test and train data split 
The cicles are the training data and the cross are the test data

```python
# we initialise a PCA object that will keep the first 2 principle components
pca = PCA(n_components = 2)
pca.fit(train_data)
# Get the Dimensionnality reduction applied on the training and testing datas
dimReducedTrainData = pca.transform(train_data)
dimReducedTestData = pca.transform(test_data)
```

```python
# Scatter the reduced dimensionality training and testing data
scatter(dimReducedTrainData, dimReducedTestData, train_labels, test_labels)
plt.title('Wine Dataset in Principle Component space')
plt.xlabel('Principle Component 0')
plt.ylabel('Principle Component 1')
plt.show()
```

### Task 3.3


```python
# create a LDA object keeping only the first 2 components
lda = LinearDiscriminantAnalysis(n_components = 2)
lda.fit(train_data, train_labels)
# applying dimensionality reduction on the train and test datas
dimReducedLDATrain = lda.transform(train_data)
dimReducedLDATest = lda.transform(test_data)
```

```python
# Scattering the LDA dimensionally reduced training and testing datas
scatter(dimReducedLDATrain, dimReducedLDATest, train_labels, test_labels)
plt.title('Wine Dataset in Linear Discriminant space')
plt.xlabel('Principle Component 0')
plt.ylabel('Principle Component 1')
plt.show()
```

##### As we can see with the 2 differents scatter plots, the PCA only reduces the dimensionnality of the datas, where LDA reduces the dimensionnality and also produces a clustering of the datas. This can be explained by the fact that LDA uses the labels to fit a model, which makes it a supervised learning model and thus can use the provided labels to cluster the data points.

### Task 3.4

```python
# Get the mean vector of the training data
mean = np.mean(train_data, axis=0)

# function to give the mean-centered data
def meanCentre(data, mean):
    size = len(data)
    result = []
    for i in range(size):
        result.append(data[i] - mean)
    return np.array(result)

meanCenteredTrain = meanCentre(train_data, mean)
```

```python
# Calculating the SVD values
u, s, vh = np.linalg.svd(meanCenteredTrain)
# slicing the right-singular vectors to get the first 2 significant eigenvectors
projection_matrix = vh[:2]
projected_train_data = meanCenteredTrain @ np.transpose(projection_matrix)
print("The shape of the mean centered train data is ",meanCenteredTrain.shape)
print("The shape of the projected mean centered train data is ",projected_train_data.shape)
```

We know that the meanCenteredTrain matrix is of shape (145x13) (aka 145 elements with 13 features.
The projection matrix is of shape (2x13), and in order to perform a matrix multiplication, it is fundamental that we have the dimension of both matrices so that the product of matrix m1 of shape (nxm) and the matrix m2 of shape (mxo) would result in a matrix of shape (nxo). The transpose of vh allows that matrix multiplication.
The dimensionality reduction can be seen by the shape of the data matrix, which went from (145,13) to (145,2)

```python
# Projection of the testing data
meanCenteredTest = meanCentre(test_data, mean)

projected_test_data = meanCenteredTest @ np.transpose(projection_matrix)
```

For projecting the test data, we still use the mean vector from the train data and the same projection matrix as the one used for the train data because we want to evaluate the data that our model will use, so we can know if he can perform well with the training data we gave to it.

```python
# visualisation of the manually projected datas
scatter(projected_train_data, projected_test_data, train_labels, test_labels)
plt.title('Wine Dataset in manual Principle Component space')
plt.xlabel('Principle Component 0')
plt.ylabel('Principle Component 1')
plt.show()
```

```python
# loading the dataset
data = np.load("Datas/task3_5_data.npy")
label = np.load('Datas/task3_5_labels.npy')
print(data.shape)
```

#### To decide the number of principle components, we need :
1. to calculate the total variation for every PCs
2. to calculate the % of the total variation for each PCs
3. add the % values of each PCs in order from highest to lowest value until we reach the 90%
4. see how many PCs we keep

```python
# perform the SVD on the dataset
svd = TruncatedSVD(n_components = 783)
svd.fit(data)
```

```python
# we get the cumulative sum of the % total variation that each PC has
sumRatio = np.cumsum(svd.explained_variance_ratio_)
# and we find where the cumulative sum gets to 90%
goal = np.where(np.cumsum(svd.explained_variance_ratio_) >= 0.90)[0][0]
print("The minimum number of PC needed to have 90% of the total variance is", goal)
```

```python
# then we plot the vector of the cumulative sum and see which number of PC gives us 90%
plt.figure()
plt.plot(sumRatio)
plt.axhline(y=0.9, color='r', linestyle='-')
plt.axvline(x = goal, color='r', linestyle='-')
plt.show()
```

#### After performing a PCA "by hand" on the dataset, we could retrieve the minimal number of Principal Components needed in order to retain at least 90% of the variance in the data.
